```python
import numpy as np
```

```python
import cv2 as cv
```

```python
img=cv.imread('/content/jigsaw.jpg')
```

```python
#Making a copy of image to not disturb the original image
copy=img.copy()
```

```python

#Cropping different parts of image to assemble again
```

```python
c1=copy[0:200,0:190]

c2=copy[200:400,0:190]

c3=copy[150:330,515:700]

c4=copy[370:421,370:796]
```

```python
#Rearranging the different parts of image
```

```python
img[0:200,0:190]=cv.flip(c2,0)

img[150:330,515:700]=cv.flip(c3,1)

img[200:400,0:190]=c1

img[370:421,370:796]=cv.flip(c4,0)
```
